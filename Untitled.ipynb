# Explotory Data analysis

# Q1
does the budget of film affect the ratings? 

```python
dfm.plot(x ='vote_average', y='budget_adj', kind = 'Scatter',figsize=(12,10));
plt.title("Voting rate against the Budget")
plt.ylabel("Budget in 100 Millions")
plt.xlabel("Average Voting Rate")

```

```python
dfm.groupby('vote_average').mean()
```

As shown in the graph above, it seems that the higher the budget the higher the rating and that makes sense as more money put into a movie means better montage .......

# Q2
does the number of voters affect the voting rate

```python
dfm.plot(x ='vote_average', y='vote_count', kind = 'Scatter',figsize=(12,10));
plt.title("Voting Rate Against Number of Voters")
plt.ylabel("Number of Voters")
plt.xlabel("Average Voting Rate")
```

As the graph above shows, it looks like as number of voters incresase the voting rate also increase

# Q3
Is the release year aaffact number of movies released?

```python
dfm['release_year'].hist(figsize=(10,8))
plt.title("Release Year against Number of Movies ")
plt.ylabel("Number of Movies Released")
plt.xlabel("Release Year")
```

As the histogram suggest, recent years have more movies released
